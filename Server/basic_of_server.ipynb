# Component required for Server



Collection of server called as data centre



```python
Application: Imagine you have a photo-sharing app like Instagram. The application includes everything users interact with, such as uploading photos, adding captions, and liking posts.

Data: In our Instagram example, the data would include the photos users upload, their captions, user profiles, and any other information stored within the app's database.

Runtime: When a user uploads a photo to Instagram, the code responsible for resizing the image and saving it to the database runs in a runtime environment. This could be the server where Instagram's backend code is executed.

Middleware: Think of middleware as the glue that connects different parts of the application together. In our Instagram app, middleware might include services for sending push notifications to users when they receive likes or comments on their posts.

Operating System (O/S): Just like your computer or phone has an operating system (like Windows, macOS, or Android), the servers running the Instagram app also have an operating system. For example, they might run on Amazon Linux, which is optimized for use on Amazon Web Services.

Virtualization: Virtualization allows us to create multiple virtual servers on a single physical server. For example, instead of having a separate physical server for each user's Instagram account, virtualization lets us run multiple Instagram accounts on the same physical server, keeping costs lower and resources more efficiently utilized.

Servers: These are like the computers that host our Instagram app. Each server runs the Instagram software and serves requests from users' devices.

Storage: Storage is where all the data in our Instagram app is kept. This includes photos, user profiles, comments, and more. Services like Amazon S3 provide scalable, reliable storage for these kinds of data.

Network: The network allows communication between different parts of our Instagram app. For example, when you upload a photo, it travels over the network to the server, where it's processed and stored. Then, when someone views your photo, it's sent back over the network to their device.
```
